```python
import pandas as pd
pd.set_option('display.max_columns', None)
import json

# --> Import functions from "process" script
import sys
sys.path.append('../src')

from process import categorise_serve_direction, get_point_level_info, get_match_point_level_info


```

```python
# -- > List of all json files with Djoker
djoko_match_list = ['year_2019_SM002_tracking_data.json',
                    'year_2019_SM032_tracking_data.json',
                    'year_2019_SM064_tracking_data.json',
                    'year_2020_SM008_tracking_data.json',
                    'year_2020_SM016_tracking_data.json',
                    'year_2019_SM004_tracking_data.json',
                    'year_2020_SM064_tracking_data.json',
                    'year_2019_SM008_tracking_data.json',
                    'year_2019_SM016_tracking_data.json',
                    'year_2020_SM004_tracking_data.json',
                    'year_2020_SM002_tracking_data.json',
                    'year_2020_SM032_tracking_data.json',
                    'year_2020_SM001_tracking_data.json']


djoko_data_frame = pd.DataFrame([])
for djoko_match_filename in djoko_match_list:
    
    with open('/Users/petertea/tennis_analytics/projects/roland_garros_tracking_data/json_data/' + djoko_match_filename) as filename:
      djoko_tracking_data_json = json.load(filename)
    
    data_to_add = get_match_point_level_info(djoko_tracking_data_json)
    data_to_add['match_id'] = djoko_match_filename
    
    year = [int(s) for s in djoko_match_filename.split('_') if s.isdigit()][0]
    data_to_add['year'] = year
    djoko_data_frame = djoko_data_frame.append( data_to_add )




```

```python
djoko_data_frame.to_csv('djokovic_roland_garros_19_20.csv', index=False)
```
